```python
import os
from PIL import Image
import mediapipe as mp
import numpy as np
import cv2
def load_images_preprocess(root_dir, image_extensions={'.jpg', '.png', '.jpeg'}):
    mp_hands = mp.solutions.hands
    hands = mp_hands.Hands(max_num_hands=1)
    image_landmarks = []
    labels = []

    for label_name in os.listdir(root_dir):
        label_path = os.path.join(root_dir, label_name)
        if not os.path.isdir(label_path):
            continue

        for filename in os.listdir(label_path):
            ext = os.path.splitext(filename)[1].lower()
            if ext in image_extensions:
                image_path = os.path.join(label_path, filename)
                img = cv2.imread(image_path)
                img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
                result = hands.process(img)
                if result.multi_hand_landmarks:
                     for hand_landmarks in result.multi_hand_landmarks:
                        h, w, _ = img.shape
                        landmarks = np.array([[lm.x,lm.y] for lm in hand_landmarks.landmark])
                        wrist = np.abs(landmarks[0])
                        print(wrist)
                        landmarks -= wrist
                image_landmarks.append(landmarks)
                labels.append(label_name)

    return image_landmarks, labels
```

```python
def show_report(y_pred,y_test,y_train_pred,y_train):
    t_acc =accuracy_score(y_train_pred, y_train) 
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average='macro')     
    rec = recall_score(y_test, y_pred, average='macro')        
    f1 = f1_score(y_test, y_pred, average='macro')             
    conf_matrix = confusion_matrix(y_test, y_pred)
    report = classification_report(y_test, y_pred)
    print(f" train Accuracy: {t_acc:.4f}")
    print(f" test Accuracy: {acc:.4f}")
    print(f"Precision (macro): {prec:.4f}")
    print(f"Recall (macro): {rec:.4f}")
    print(f"F1 Score (macro): {f1:.4f}")
    print("Confusion Matrix:")
    plt.figure(figsize=(10, 8))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=met, yticklabels=met)
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title('Confusion Matrix')
    plt.tight_layout()
    plt.show()
    print("\nClassification Report:")
    print(report)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
X_flat = x.reshape(x.shape[0], -1)
X_train, X_test, y_train, y_test = train_test_split(X_flat, y, test_size=0.2,random_state=42)
```

```python
log_model = LogisticRegression(multi_class='multinomial', solver='lbfgs',max_iter=1000)
log_model.fit(X_train , y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

y_pred = log_model.predict(X_test)
y_train_pred = log_model.predict(X_train)
show_report(y_pred,y_test,y_train_pred,y_train)
```

```python
from sklearn.svm import SVC
svm_model = SVC(kernel='rbf') 
svm_model.fit(X_train, y_train)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras import regularizers
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
model = Sequential([
    Dense(256, activation='relu', input_shape=(42,)),
    Dense(128, activation='relu',kernel_regularizer=regularizers.l2(0.01)),
    Dense(64,activation = 'relu',kernel_regularizer=regularizers.l2(0.01)),
    Dense(len(met), activation='softmax')
])
X_train, X_test, y_train, y_test = train_test_split(X_flat, y, test_size=0.2, random_state=42)
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(X_train, y_train, epochs=1000, batch_size=32, validation_split=0.2,callbacks=[early_stop])
```

```python
y_pred_probs = model.predict(X_test)
y_pred = y_pred_probs.argmax(axis=1)

if len(y_test.shape) == 2 and y_test.shape[1] > 1:
    y_test_labels = y_test.argmax(axis=1)
else:
    y_test_labels = y_test

print(classification_report(y_test_labels, y_pred))
```

```python
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout
conv_model = Sequential([
    Conv1D(128, kernel_size=3, activation='relu', input_shape=(21, 2)),
    MaxPooling1D(pool_size=2),

    Conv1D(64, kernel_size=3, activation='relu'),
    MaxPooling1D(pool_size=2),
    Flatten(),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(len(met), activation='softmax')
])
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
conv_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
conv_model.fit(X_train, y_train, epochs=1000, batch_size=32, validation_split=0.2)
```

```python
y_pred_probs = conv_model.predict(X_test)
y_pred = y_pred_probs.argmax(axis=1)

if len(y_test.shape) == 2 and y_test.shape[1] > 1:
    y_test_labels = y_test.argmax(axis=1)
else:
    y_test_labels = y_test

print(classification_report(y_test_labels, y_pred))
```
